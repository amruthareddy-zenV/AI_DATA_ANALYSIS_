### Data Drift: Visualizing Concept Drift
**Description**: Simulate concept drift by gradually changing the distribution of a binary target variable over simulated time periods. Plot these changes.

```python
# write your code from here
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

# Simulate data with concept drift
np.random.seed(42)
time_periods = 100
data = {'Time': np.arange(time_periods),
    'Target': np.where(np.arange(time_periods) < time_periods / 2,
               np.random.binomial(1, 0.3, time_periods),
               np.random.binomial(1, 0.7, time_periods))}

df = pd.DataFrame(data)

# Plot the concept drift
rolling_mean = df['Target'].rolling(window=10).mean()
plt.figure(figsize=(10, 6))
plt.plot(df['Time'], rolling_mean, label='Rolling Mean (Window=10)', color='blue')
plt.scatter(df['Time'], df['Target'], alpha=0.5, label='Target', color='orange')
plt.title('Simulating Concept Drift')
plt.xlabel('Time')
plt.ylabel('Target')
plt.legend()
plt.grid()
plt.show()
```
