```python
# Advanced Seaborn Plot - Pairplot
# Objective: Use Seaborn to create a pairplot to visualize pairwise relationships in a dataset.
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
iris = sns.load_dataset('iris')

# Customize plot style
sns.set_style("whitegrid", {'grid.linestyle': ':'})
plt.figure(figsize=(12, 12))

# Create pairplot with advanced customization
pair_grid = sns.pairplot(
    data=iris,
    hue="species",                # Color by species
    palette="viridis",            # Modern color palette
    height=2.5,                   # Size of each subplot
    aspect=1.2,                   # Width/height ratio
    diag_kind="kde",              # Diagonal plots show density
    plot_kws={
        's': 40,                  # Point size
        'edgecolor': 'black',     # Point borders
        'linewidth': 0.5,         # Border thickness
        'alpha': 0.8              # Transparency
    },
    diag_kws={
        'fill': True,             # Fill under density curve
        'alpha': 0.3              # Transparency for fill
    },
    corner=False                  # Show both upper/lower triangles
)

# Add super title
pair_grid.fig.suptitle('Iris Dataset Pairwise Relationships', 
                      y=1.02, 
                      fontsize=16,
                      fontweight='bold')

# Adjust layout
plt.tight_layout()
plt.show()
```
