```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

import matplotlib.pyplot as plt

# Simulated performance metrics over time
time_periods = ['Week 1', 'Week 2', 'Week 3', 'Week 4']
accuracy = [0.95, 0.93, 0.90, 0.88]
precision = [0.94, 0.92, 0.89, 0.87]
recall = [0.96, 0.94, 0.91, 0.89]

# Plotting the metrics
plt.figure(figsize=(10, 6))
plt.plot(time_periods, accuracy, label='Accuracy', marker='o')
plt.plot(time_periods, precision, label='Precision', marker='o')
plt.plot(time_periods, recall, label='Recall', marker='o')

plt.title('Model Performance Metrics Over Time')
plt.xlabel('Time Periods')
plt.ylabel('Metric Value')
plt.ylim(0.8, 1.0)
plt.legend()
plt.grid(True)
plt.show()
```
