```python
# Customizing Matplotlib Plot Styles
# Description: Customize the Matplotlib bar plot from Question 1 with a grid, different color, and custom fonts.
import matplotlib.pyplot as plt
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]
plt.bar(products,sales,color = 'darkgreen')
plt.grid(axis = 'y', linestyle = '--', alpha = 0.6)
plt.title('Product Sales (in Thousands)', fontsize=16, fontweight='bold', fontfamily='serif')
plt.xlabel('Products', fontsize=12, fontfamily='monospace')
plt.ylabel('Sales (in Thousands)', fontsize=12, fontfamily='monospace')
plt.show()
```
