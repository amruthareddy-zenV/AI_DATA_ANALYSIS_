```python
# Scatter Plot of Variable Relationships
# Description: Use Seaborn to create a scatter plot showing the relationship between two
# variables.
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset (tips dataset built into Seaborn)
data = sns.load_dataset("tips")

# Set up the visualization
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid", {'grid.color': '.9'})
sns.set_palette("rocket")

# Create scatter plot with enhanced features
scatter = sns.scatterplot(
    data=data,
    x="total_bill",
    y="tip",
    hue="size",        # Color by party size
    size="size",       # Size by party size
    sizes=(30, 200),   # Range of point sizes
    alpha=0.8,         # Slight transparency
    edgecolor="black", # Black borders on points
    linewidth=0.5      # Border thickness
)

# Add regression line
sns.regplot(
    data=data,
    x="total_bill",
    y="tip",
    scatter=False,     # Don't overlay scatter points
    ci=95,             # 95% confidence interval
    line_kws={
        'color': 'black',
        'linestyle': '--',
        'alpha': 0.6
    }
)

# Customize titles and labels
plt.title("Relationship Between Total Bill and Tips", 
         fontsize=14, pad=15)
plt.xlabel("Total Bill ($)", fontsize=12)
plt.ylabel("Tip Amount ($)", fontsize=12)

# Format currency on axes
plt.gca().xaxis.set_major_formatter('${x:.0f}')
plt.gca().yaxis.set_major_formatter('${x:.0f}')

# Improve legend
plt.legend(
    title="Party Size",
    bbox_to_anchor=(1.05, 1),
    borderpad=1,
    frameon=True
)

# Add correlation annotation
corr = data['total_bill'].corr(data['tip'])
plt.text(0.95, 0.1, 
        f'Pearson r = {corr:.2f}',
        transform=plt.gca().transAxes,
        ha='right',
        bbox=dict(facecolor='white', alpha=0.8))

# Adjust layout
plt.tight_layout()
plt.show()
```
