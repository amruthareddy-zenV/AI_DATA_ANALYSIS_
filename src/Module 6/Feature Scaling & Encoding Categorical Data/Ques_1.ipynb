```python
# Question 1: Feature Scaling
# Task: Explain why feature scaling is essential and demonstrate the impact of unscaled features on a machine learning model.




# Question 2: Min-Max Scaling
# Task: Implement Min-Max Scaling on the Iris dataset.



# Question 3: Standardization (Z-score Scaling)
# Task: Implement Standardization using Z-score scaling on the Iris dataset.




# Question 4: Robust Scaling
# Task: Implement Robust Scaling to handle outliers in the Iris dataset.

# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Load the Iris dataset
iris = load_iris()
X = iris.data
feature_names = iris.feature_names

# Create a Pandas DataFrame for better visualization (optional)
# This helps in displaying the data with feature names
df = pd.DataFrame(X, columns=feature_names)

print("Original Data (first 5 rows):")
print(df.head())
print("\nOriginal Data Description:")
# The describe() method provides summary statistics like mean, std, min, max
print(df.describe())

# Initialize the MinMaxScaler
# This scaler transforms features by scaling each feature to a given range.
# The default range is [0, 1].
scaler = MinMaxScaler()

# Apply Min-Max Scaling to the data
# fit_transform calculates the min and max for each feature and then
# applies the scaling transformation.
X_scaled_minmax = scaler.fit_transform(X)

# Create a DataFrame for the scaled data
# This helps in viewing the scaled data in a structured format
df_scaled_minmax = pd.DataFrame(X_scaled_minmax, columns=feature_names)

print("\nMin-Max Scaled Data (first 5 rows):")
print(df_scaled_minmax.head())
print("\nMin-Max Scaled Data Description:")
# Notice the min and max values are now within the [0, 1] range
print(df_scaled_minmax.describe())

# You can now use X_scaled_minmax for training your machine learning model
# Algorithms sensitive to feature scale, like KNN or SVM, will benefit from this.




```
