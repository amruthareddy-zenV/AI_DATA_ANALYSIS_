```python
# Seaborn Histogram with KDE
# Description: Create a histogram using Seaborn that includes a Kernel Density Estimate (KDE) for marks
# data: [23, 45, 56, 78, 89, 90, 67, 88, 76, 59].
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Data
marks = [23, 45, 56, 78, 89, 90, 67, 88, 76, 59]

# Set style
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

# Create histogram with KDE
ax = sns.histplot(
    data=marks,
    bins=5,
    kde=True,
    color='royalblue',
    edgecolor='black',
    linewidth=1.2,
    alpha=0.7,
    stat='density'  # Shows density instead of count
)

# Customize KDE line
kde_line = ax.lines[0]
kde_line.set_color('darkred')
kde_line.set_linewidth(2.5)
kde_line.set_linestyle('--')

# Add titles and labels
plt.title('Distribution of Student Marks with KDE', 
          fontsize=16, pad=20, fontweight='bold')
plt.xlabel('Marks', fontsize=14)
plt.ylabel('Density', fontsize=14)

# Customize ticks
plt.xticks(np.arange(0, 101, 10))
plt.yticks(fontsize=12)

# Add mean line
mean_mark = np.mean(marks)
ax.axvline(mean_mark, color='green', linestyle=':', linewidth=2.5)
ax.text(mean_mark+2, 0.02, f'Mean: {mean_mark:.1f}', 
        color='green', fontsize=12)

# Add annotation
ax.text(0.95, 0.95, 'Kernel Density Estimate',
        transform=ax.transAxes,
        color='darkred',
        fontsize=12,
        ha='right')

# Remove top and right spines
sns.despine()

plt.tight_layout()
plt.show()
```
