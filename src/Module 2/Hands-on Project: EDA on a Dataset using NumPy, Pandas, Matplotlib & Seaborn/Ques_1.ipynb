```python
# Title: Load & Examine a Basic Dataset
# Description: Load an example dataset such as the Iris dataset and perform basic exploration.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

# Basic dataset examination
print("=== Dataset Overview ===")
print(f"Shape: {iris_df.shape}")
print("\nFirst 5 rows:")
print(iris_df.head())

print("\n=== Basic Statistics ===")
print(iris_df.describe())

print("\n=== Class Distribution ===")
print(iris_df['species'].value_counts())

# Visual exploration
plt.figure(figsize=(15, 10))

# Pairplot for multivariate relationships
plt.subplot(2, 2, 1)
sns.pairplot(iris_df, hue='species', palette='viridis')
plt.suptitle('Iris Dataset Pairwise Relationships', y=1.02)

# Boxplot for feature distribution by species
plt.subplot(2, 2, 2)
sns.boxplot(data=iris_df, x='species', y='sepal length (cm)', palette='Set2')
plt.title('Sepal Length Distribution by Species')

# Correlation heatmap
plt.subplot(2, 2, 3)
corr_matrix = iris_df.iloc[:, :4].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')

# Violin plot showing feature distributions
plt.subplot(2, 2, 4)
sns.violinplot(data=iris_df.iloc[:, :4], inner='quartile', palette='pastel')
plt.title('Feature Distribution Overview')

plt.tight_layout()
plt.show()

# Advanced statistics
print("\n=== Group-wise Statistics ===")
print(iris_df.groupby('species').describe().transpose())

# Missing values check
print("\n=== Missing Values ===")
print(iris_df.isnull().sum())
```
