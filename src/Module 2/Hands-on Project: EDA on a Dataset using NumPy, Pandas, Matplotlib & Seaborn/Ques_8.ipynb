```python
# Title: Advanced Data Visualization
# Description: Create a pair plot for the dataset to display pairwise relationships.
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset (replace with your actual data)
df = sns.load_dataset('iris')

# Create pair plot with additional customization
pair_plot = sns.pairplot(
    data=df,
    hue='species',             # Color by species
    palette='viridis',         # Color scheme
    diag_kind='kde',          # Diagonal shows KDE plots
    plot_kws={'alpha': 0.8, 's': 30},  # Point transparency and size
    corner=False              # Show full matrix (set to True for triangular)
)

# Add title and adjust layout
pair_plot.fig.suptitle('Pairwise Relationships by Species', y=1.02)
plt.tight_layout()
plt.show()
```
