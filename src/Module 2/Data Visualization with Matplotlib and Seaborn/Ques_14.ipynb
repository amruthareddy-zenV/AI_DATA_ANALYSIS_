```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x) * 2 + 3
y2 = np.cos(x) * 1.5 + 2
y3 = np.exp(-x/3) * 4

# Set up the figure with custom style
plt.figure(figsize=(10, 6), facecolor='#f8f9fa')  # Light gray background
ax = plt.axes(facecolor='white')  # White plot area

# Customize grid and spines
ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.7, color='gray')
for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)
for spine in ['left', 'bottom']:
    ax.spines[spine].set_color('gray')

# Plot lines with custom styles
line1, = ax.plot(x, y1, 
                color='#3498db',  # Blue
                linewidth=2.5,
                linestyle='-',
                marker='o',
                markersize=6,
                markevery=5,
                markerfacecolor='white',
                markeredgecolor='#3498db',
                markeredgewidth=1.5,
                label='Production')

line2, = ax.plot(x, y2, 
                color='#e74c3c',  # Red
                linewidth=2,
                linestyle='--',
                dash_capstyle='round',
                label='Consumption')

line3, = ax.plot(x, y3, 
                color='#2ecc71',  # Green
                linewidth=3,
                linestyle='-.',
                alpha=0.8,
                label='Inventory')

# Add titles and labels
plt.title('Business Metrics Over Time', 
         fontsize=16, pad=20, 
         fontweight='bold', 
         color='#2c3e50')
plt.xlabel('Time (weeks)', fontsize=12, labelpad=10, color='#34495e')
plt.ylabel('Metric Value', fontsize=12, labelpad=10, color='#34495e')

# Customize ticks
plt.xticks(fontsize=10, color='#7f8c8d')
plt.yticks(fontsize=10, color='#7f8c8d')

# Add legend with shadow
legend = plt.legend(frameon=True, 
                   framealpha=1,
                   shadow=True,
                   borderpad=1,
                   facecolor='white',
                   edgecolor='#bdc3c7')
legend.get_frame().set_linewidth(0.5)

# Add annotation for peak value
peak_x = x[np.argmax(y1)]
peak_y = max(y1)
ax.annotate(f'Peak: {peak_y:.1f}',
           xy=(peak_x, peak_y),
           xytext=(peak_x+1, peak_y-0.5),
           arrowprops=dict(arrowstyle='->', color='#3498db'),
           bbox=dict(boxstyle='round,pad=0.5', fc='white', ec='#3498db', alpha=0.8))

# Add light watermark
plt.text(0.95, 0.05, 'CONFIDENTIAL',
        transform=ax.transAxes,
        fontsize=40,
        color='gray',
        alpha=0.1,
        ha='right',
        va='bottom')

plt.tight_layout()
plt.show()
```
