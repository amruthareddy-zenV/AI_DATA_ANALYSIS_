```python
import numpy as np

# Task 2.1: Visualize Data Distribution with Histograms

import matplotlib.pyplot as plt

# Example data for visualization
data = np.random.normal(loc=50, scale=10, size=1000)

# Plot histogram
plt.hist(data, bins=30, color='blue', alpha=0.7)
plt.title('Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Task 6: Handling Outliers with Capping

# Define the lower and upper bounds for capping
lower_bound = np.percentile(data, 5)  # 5th percentile
upper_bound = np.percentile(data, 95)  # 95th percentile

# Apply capping
data_capped = np.clip(data, lower_bound, upper_bound)

# Display the bounds and a sample of the capped data
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print("Sample of capped data:", data_capped[:10])
```
