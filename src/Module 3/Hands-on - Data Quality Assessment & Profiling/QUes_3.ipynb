```python
# Part 1: Load a Dataset & Check Missing Values

# Task 1: Iris Dataset
# - Load the dataset using Pandas.
# - Check for missing values in the entire dataset.








# Part 2: Identify Duplicates & Inconsistencies

# Task 2: Inconsistent Entries in a Sample Dataset
# - Assume you have a dataset with a 'Gender' column. Identify inconsistent entries like 'M', 'Male', or 'male'.









# Part 3: Generate a Data Quality Report

# Task 3: Iris Dataset Summary
# - Generate basic descriptive statistics for the Iris dataset.
import pandas as pd
import numpy as np

# Part 1: Load Iris Dataset and Check Missing Values
# ------------------------------------------------
print("="*50)
print("PART 1: LOADING IRIS DATASET & MISSING VALUES CHECK")
print("="*50)

# Load the Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
iris_df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                      columns=iris['feature_names'] + ['target'])
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Check for missing values
missing_values = iris_df.isnull().sum()
print("\nMissing Values Check:")
print(missing_values)

# Part 2: Identify Inconsistent Entries (Sample Gender Column)
# ----------------------------------------------------------
print("\n" + "="*50)
print("PART 2: INCONSISTENT ENTRIES DEMONSTRATION")
print("="*50)

# Create sample dataset with inconsistent gender entries
gender_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Gender': ['F', 'Male', 'female', 'M', 'FEMALE']
}
gender_df = pd.DataFrame(gender_data)

# Identify inconsistent gender entries
print("\nOriginal Gender Column:")
print(gender_df['Gender'].value_counts())

# Standardize gender entries
gender_df['Gender_clean'] = gender_df['Gender'].str.lower().str[0]
print("\nStandardized Gender Column:")
print(gender_df[['Name', 'Gender', 'Gender_clean']])

# Part 3: Generate Iris Dataset Summary
# -----------------------------------
print("\n" + "="*50)
print("PART 3: IRIS DATASET SUMMARY REPORT")
print("="*50)

def generate_iris_report(df):
    """Generate a comprehensive report for the Iris dataset"""
    
    report = {
        'dataset_info': {
            'Total Samples': len(df),
            'Total Features': len(df.columns),
            'Missing Values': df.isnull().sum().sum(),
            'Species Distribution': df['species'].value_counts().to_dict()
        },
        'descriptive_stats': df.describe(),
        'species_stats': df.groupby('species').describe().stack()
    }
    return report

# Generate and display report
iris_report = generate_iris_report(iris_df)

# Print basic information
print("\nDATASET INFORMATION:")
for k, v in iris_report['dataset_info'].items():
    print(f"{k:<20}: {v}")

# Print descriptive statistics
print("\n\nDESCRIPTIVE STATISTICS:")
print(iris_report['descriptive_stats'])

# Print species-specific statistics
print("\n\nSPECIES-SPECIFIC STATISTICS:")
print(iris_report['species_stats'])



```
