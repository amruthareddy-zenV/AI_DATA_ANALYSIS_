**Task**: Hierarchical Clustering for Deduplication

**Steps**:
1. Data Set: Obtain a dataset containing duplicate employee information.
2. Perform Clustering: Use hierarchical agglomerative clustering to cluster the employee
records.
3. Evaluate Duplicates: Determine duplicates by analyzing the clusters formed.
4. Clean Data: Remove duplicate employee records found during clustering.

```python
%pip install scipy

import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram

import matplotlib.pyplot as plt

# Sample dataset with duplicate employee information
data = {
    'EmployeeID': [1, 2, 3, 4, 5, 6],
    'Name': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob', 'Charlie'],
    'Age': [25, 30, 25, 35, 30, 35],
    'Department': ['HR', 'IT', 'HR', 'Finance', 'IT', 'Finance']
}

# Create a DataFrame
df = pd.DataFrame(data)

# Encode categorical data (e.g., Name and Department) for clustering
df_encoded = pd.get_dummies(df[['Name', 'Department']])

# Standardize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_encoded)

# Perform hierarchical clustering
linkage_matrix = linkage(df_scaled, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(linkage_matrix, labels=df['EmployeeID'].values)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Employee ID')
plt.ylabel('Distance')
plt.show()
```
