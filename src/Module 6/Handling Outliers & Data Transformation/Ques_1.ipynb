```python
# Question: Visualizing Outliers Using Boxplots
# Description: Use a boxplot to visually detect outliers in a dataset.

# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
X = iris.data
feature_names = iris.feature_names

# Create a Pandas DataFrame for easier plotting with Seaborn
df = pd.DataFrame(X, columns=feature_names)

print("Original Data (first 5 rows):")
print(df.head())
print("\nOriginal Data Description:")
print(df.describe())

# --- Create Boxplots to Visualize Outliers ---

# Set the style for the plot
sns.set_style("whitegrid")

# Create boxplots for each feature
# The 'data' parameter specifies the DataFrame to use
# The 'orient="h"' parameter can be used for horizontal boxplots if preferred
plt.figure(figsize=(10, 6)) # Set the figure size for better readability
sns.boxplot(data=df, orient="v", palette="Set2") # 'v' for vertical orientation

# Add title and labels
plt.title('Boxplots of Iris Dataset Features to Visualize Outliers')
plt.ylabel('Measurement (cm)') # Assuming units are cm for Iris features
plt.xlabel('Features')

# Rotate x-axis labels if needed for better readability
plt.xticks(rotation=45, ha='right')

# Adjust layout to prevent labels overlapping
plt.tight_layout()

# Display the plot
plt.show()

# --- Interpretation ---
# In the generated boxplots:
# - The box represents the interquartile range (IQR), from the first quartile (Q1) to the third quartile (Q3).
# - The line inside the box is the median (Q2).
# - The "whiskers" extend from the box to the minimum and maximum values within 1.5 times the IQR.
# - Points plotted individually beyond the whiskers are typically considered potential outliers.
# Look for individual points outside the whiskers for each feature to identify outliers.

```
