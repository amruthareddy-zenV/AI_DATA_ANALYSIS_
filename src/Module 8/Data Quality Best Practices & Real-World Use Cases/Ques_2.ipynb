```python
import pandas as pd

import matplotlib.pyplot as plt

# Simulate historical data for data quality KPIs
data = {
    'Date': pd.date_range(start='2023-01-01', periods=12, freq='M'),
    'Completeness': [95, 96, 94, 97, 96, 95, 94, 96, 97, 98, 99, 97],
    'Accuracy': [92, 93, 91, 94, 93, 92, 91, 93, 94, 95, 96, 94],
    'Timeliness': [90, 91, 89, 92, 91, 90, 89, 91, 92, 93, 94, 92]
}

# Create a DataFrame
df_trends = pd.DataFrame(data)

# Plot trends for KPIs
plt.figure(figsize=(10, 6))
for kpi in ['Completeness', 'Accuracy', 'Timeliness']:
    plt.plot(df_trends['Date'], df_trends[kpi], label=kpi)

plt.title('Data Quality Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Percentage')
plt.legend()
plt.grid(True)
plt.show()
```
