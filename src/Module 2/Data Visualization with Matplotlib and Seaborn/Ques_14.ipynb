```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.
import matplotlib.pyplot as plt
x=[0,1,2,3,4,5]
y=[0,1,4,9,16,25]
plt.plot(x,y,color='green',linestyle='-',linewidth=2,marker='o',markersize=8,markerfacecolor='red')
plt.title('Customized Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
