```python
# Basic Seaborn Scatter Plot
# Description: Using Seaborn, recreate the scatter plot from Question 2 for height vs. weight.
import seaborn as sns
import matplotlib.pyplot as plt

# Data
heights = [160, 170, 165, 180, 175]  # in centimeters
weights = [55, 65, 60, 80, 76]       # in kilograms

# Set Seaborn style
sns.set(style="whitegrid", context="notebook")  # Clean style with grid

# Create scatter plot with regression line
plt.figure(figsize=(8, 6))
ax = sns.regplot(x=heights, y=weights, 
                 scatter_kws={'s': 100, 'color': 'darkblue', 'alpha': 0.8},
                 line_kws={'color': 'red', 'linestyle': '--'})

# Add titles and labels
plt.title('Height vs Weight Correlation', fontsize=14, pad=20)
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Weight (kg)', fontsize=12)

# Add data point labels
for i, (h, w) in enumerate(zip(heights, weights)):
    ax.text(h+1, w-1, f'({h},{w})', fontsize=10)

# Adjust axis limits
plt.xlim(155, 185)
plt.ylim(50, 85)

# Remove top and right spines
sns.despine()

plt.tight_layout()
plt.show()
```
