```python
# Title: Basic Data Visualization
# Description: Plot the distribution of one feature using Seaborn.
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset (replace with your actual data)
data = sns.load_dataset('iris')

# Plot distribution of one feature (sepal_length in this example)
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='sepal_length', kde=True)

# Add titles and labels
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
